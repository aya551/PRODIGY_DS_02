```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv(r"C:\Users\ayael\Downloads\titanic\train.csv")
# Check and remove duplicates
duplicates = data.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

```python
print("\nMissing values in the dataset:")
print(data.isnull().sum())
```

```python
# Handling missing 'Age' values
median_age = data['Age'].median()
data['Age'].fillna(median_age, inplace=True)
print(f"Median Age used for missing values: {median_age}")
```

```python
#Handling missing 'Embarked' values
most_common_embarked = data['Embarked'].mode()[0]
data['Embarked'].fillna(most_common_embarked, inplace=True)
print(f"Most common Embarked value used: {most_common_embarked}")
```

```python
#Dropping the 'Cabin' column due to a large number of missing values
data.drop(columns=['Cabin'], inplace=True)
```

```python
#checking on Data
print(data.isnull().sum())
print(data.head())
```

```python

# Understand the dataset with descriptive statistics
print("Summary statistics of the dataset:")
print(data.describe())
```

```python
#  Distribution of  (Survived)
plt.figure(figsize=(6, 4))
sns.countplot(data=data, x='Survived')
plt.title('Survival Count')
plt.show()
```

```python
#Survival Rate by Sex
plt.figure(figsize=(6, 4))
sns.barplot(data=data, x='Sex', y='Survived')
plt.title('Survival Rate by Sex')
plt.show()
```

```python
# Survival Rate by Pclass
plt.figure(figsize=(6, 4))
sns.barplot(data=data, x='Pclass', y='Survived')
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python

# Visualizing the distribution of 'Age' and 'Fare'
plt.figure(figsize=(12,5))

plt.subplot(1, 2, 1)
sns.histplot(data['Age'], kde=True, bins=30, color='blue')
plt.title('Age Distribution')

plt.subplot(1, 2, 2)
sns.histplot(data['Fare'], kde=True, bins=30, color='green')
plt.title('Fare Distribution')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))  
# 1st plot: Fare vs Survival
plt.subplot(1, 2, 1)  
sns.histplot(data=data, x='Fare', hue='Survived', multiple="stack", kde=True)
plt.title('Fare vs Survival')
# 2nd plot: Age vs Survival
plt.subplot(1, 2, 2) 
sns.histplot(data=data, x='Age', hue='Survived', multiple="stack", kde=True)
plt.title('Age vs Survival')
plt.tight_layout()
plt.show()
```

```python
# Survival Rate by Embarked
plt.figure(figsize=(6, 4))
sns.countplot(data=data, x='Embarked', hue='Survived')
plt.title('Survival Rate by Embarked Location')
plt.show()
```

```python
# SibSp vs Survival
plt.figure(figsize=(6, 4))
sns.barplot(data=data, x='SibSp', y='Survived',color='green')
plt.title('Survival Rate by SibSp')
plt.show()
```

```python
# Parch  vs Survival
plt.figure(figsize=(6, 4))
sns.barplot(data=data, x='Parch', y='Survived')
plt.title('Survival Rate by Parch')
plt.show()
```

```python
numeric_df = data.drop(columns=['Name', 'Sex', 'Ticket', 'Embarked'])
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python

numerical_cols = ['Age', 'Fare', 'SibSp', 'Parch']
sns.pairplot(data[numerical_cols + ['Survived']], hue='Survived', palette='coolwarm', diag_kind='kde')
plt.suptitle('Pairplot of Numerical Variables vs Survival', y=1.02)
plt.show()
```
